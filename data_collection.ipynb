```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
browser = webdriver.Chrome('./chromedriver.exe')
```

```python
def linkedIN_login(browser):
    browser.get("https://www.linkedin.com/")
    user_name = 'ayonathan0@gmail.com'
    pw = '0Nward2thekitchen'
    time.sleep(5)
    user_css = '#session_key'
    pw_css = '#session_password'
    try:
        ur_elem = browser.find_element(By.CSS_SELECTOR, user_css)
        ur_elem.clear()
        ur_elem.send_keys(user_name)

        pw_elem = browser.find_element(By.CSS_SELECTOR, pw_css)
        pw_elem.clear()
        pw_elem.send_keys(pw)
        pw_elem.send_keys(Keys.RETURN)
    except:
        return None

def Collect_info(lead):
    soup = BeautifulSoup(browser.page_source, 'html.parser')

    name = 'mt2 relative'
    pos = 'text-body-medium break-words'
    expe = 'pv-text-details__right-panel'
    add = 'pv-text-details__left-panel mt2'

    search_result = soup.find_all('div', class_= name)
    if len(search_result) == 0:
        name_ = lead
        position_company = 'No Entry Found'
        expe_ = 'No Entry Found'
        addr = 'No Entry Found'
    else:

        if search_result[0].find('h1') is not None:
            name_ = search_result[0].find('h1').text
        else:
            name_ = lead

        if search_result[0].find('div', class_=pos) is not None:
            position_company = search_result[0].find('div', class_=pos).text
        else:
            position_company = ' '

        if search_result[0].find('ul', class_=expe) is not None:
            expe_ = search_result[0].find('ul', class_=expe).text
        else:
            expe_ = ' '

        if search_result[0].find('div', class_=add) is not None:
            addr = search_result[0].find('div', class_=add).text
        else:
            addr = ' '

    return name_, position_company, expe_, addr

def search(_):
    search_bar_css = '#global-nav-typeahead > input'
    Search_elem = browser.find_element(By.CSS_SELECTOR, search_bar_css)
    Search_elem.clear()
    Search_elem.send_keys(_)
    Search_elem.send_keys(Keys.RETURN)
```

```python
linkedIN_login(browser)
```

```python
prospect_df = pd.read_excel('PROSPECTS GLOBAL - Sponsor Targets Worksheet.xlsx')
prospect_df['Job Title'].replace(np.nan, ' ', inplace=True)
prospect_df['Last name'].replace(np.nan, ' ', inplace=True)
f_name = prospect_df['First name'].values.tolist()
l_name = prospect_df['Last name'].values.tolist()
organisation = prospect_df['Organisation'].values.tolist()
job_title = prospect_df['Job Title'].values.tolist()
s_term = []
for i in range(prospect_df.shape[0]):
    search_term = f_name[i] + ' ' + l_name[i] + ' ' + organisation[i] 
    s_term.append(search_term)
s_term
```

```python
Collect_info(lead='h')
```

```python
name_list = []
exper_list = []
pos_list = []
addre_list = []
count = 0
```

```python
for _ in s_term[:100]:
    time.sleep(5)
    search(_)
    time.sleep(10)
    try:
        ser_result_css = '#main > div > div'
        Search_result = browser.find_element(By.CSS_SELECTOR, ser_result_css)
        Search_result.find_element(By.TAG_NAME, 'a').click()
        time.sleep(10)
        n, p, e, a = Collect_info(_)
        name_list.append(n)
        exper_list.append(e)
        pos_list.append(p)
        addre_list.append(a)
    except:
        pass
data = {}
data['name'] = name_list
data['expe'] = exper_list
data['position_company'] = pos_list
data['addr'] = addre_list
df = pd.DataFrame(data)
df.to_csv(f'aaAdd__0_{count}.csv')
count = count + 1
```
